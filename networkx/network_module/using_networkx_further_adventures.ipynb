## A graph from Rogets thesaurus

The next cell contains some code for building a graph for Roget's thesaurus, an
example taken from the networkx site, which uses Knuth's Roget example from the
Stanford Graph Base.

```python
from __future__ import print_function
__author__ = """Brendt Wohlberg\nAric Hagberg (hagberg@lanl.gov)"""
__date__ = "$Date: 2005-04-01 07:56:22 -0700 (Fri, 01 Apr 2005) $"
__credits__ = """"""
__revision__ = ""
#    Copyright (C) 2004 by
#    Aric Hagberg <hagberg@lanl.gov>
#    Dan Schult <dschult@colgate.edu>
#    Pieter Swart <swart@lanl.gov>
#    All rights reserved.
#    BSD license.

import re
import sys

def roget_graph():
    """ Return the thesaurus graph from the roget.dat example in
    the Stanford Graph Base.
    """
    # open file roget_dat.txt.gz (or roget_dat.txt)
    #import gzip
    #fh=gzip.open('roget_dat.txt.gz','r')
    fh = open('roget.dat','r')

    G=nx.DiGraph()

    for line in fh.readlines():
        #line = line.decode()
        if line.startswith("*"): # skip comments
            continue
        if line.startswith(" "): # this is a continuation line, append
            line=oldline+line
        if line.endswith("\\\n"): # continuation line, buffer, goto next
            oldline=line.strip("\\\n")
            continue

        (headname,tails)=line.split(":")

        # head
        numfind=re.compile("^\d+") # re to find the number of this word
        head=numfind.findall(headname)[0] # get the number

        G.add_node(head)

        for tail in tails.split():
            if head==tail:
                print("skipping self loop",head,tail, file=sys.stderr)
            G.add_edge(head,tail)

    return G
```

```python
G=roget_graph()
print("Loaded roget_dat.txt containing 1022 categories.")
print("digraph has %d nodes with %d edges"\
          %(nx.number_of_nodes(G),nx.number_of_edges(G)))
UG=G.to_undirected()
print(nx.number_connected_components(UG),"connected components")
```

```python
components = list(nx.connected_components(UG))
```

```python
print(len(components))
g0 = components[0]
g1 = components[1]
```

```python
print(len(g1))
g0_graph = nx.subgraph(UG,g0)
nx.draw_spring(g0_graph)
```

## Plotting graphs with prepositioned data

The following example is due to Aric Hagberg, and is taken from the networkx gallery of examples.
It illustrates how to use networkx with prepositioned data, using an example from
Knuth's Stanford Graph Base.

```python
"""
An example using networkx.Graph().

miles_graph() returns an undirected graph over the 128 US cities from
the datafile miles_dat.txt. The cities each have location and population
data.  The edges are labeled with the distance betwen the two cities.

This example is described in Section 1.1 in Knuth's book [1,2].

References.
-----------

[1] Donald E. Knuth,
    "The Stanford GraphBase: A Platform for Combinatorial Computing",
    ACM Press, New York, 1993.
[2] http://www-cs-faculty.stanford.edu/~knuth/sgb.html


"""
__author__ = """Aric Hagberg (hagberg@lanl.gov)"""
#    Copyright (C) 2004-2006 by
#    Aric Hagberg <hagberg@lanl.gov>
#    Dan Schult <dschult@colgate.edu>
#    Pieter Swart <swart@lanl.gov>
#    All rights reserved.
#    BSD license.

import networkx as nx


def miles_graph(miles_file):
    """ Return the citees example graph in miles.dat
        from the Stanford GraphBase.
    """
    # open file miles.dat

    fh = open(miles_file,'r')

    G=nx.Graph()
    G.position={}
    G.population={}

    cities=[]
    for line in fh.readlines():
        #line = line.decode()
        if line.startswith("*"): # skip comments
            continue

        numfind=re.compile("^\d+")

        if numfind.match(line): # this line is distances
            dist=line.split()
            for d in dist:
                G.add_edge(city,cities[i],weight=int(d))
                i=i+1
        else: # this line is a city, position, population
            i=1
            (city,coordpop)=line.split("[")
            cities.insert(0,city)
            (coord,pop)=coordpop.split("]")
            (y,x)=coord.split(",")

            G.add_node(city)
            # assign position - flip x axis for matplotlib, shift origin
            G.position[city]=(-int(x)+7500,int(y)-3000)
            G.population[city]=float(pop)/1000.0
    return G

if __name__ == '__main__':
    import networkx as nx
    import re
    import sys

    G=miles_graph('miles.dat')

    print("Loaded miles.dat containing 128 cities.")
    print("digraph has %d nodes with %d edges"\
          %(nx.number_of_nodes(G),nx.number_of_edges(G)))


    # make new graph of cites, edge if less then 300 miles between them
    H=nx.Graph()
    for v in G:
        H.add_node(v)
    for (u,v,d) in G.edges(data=True):
        if d['weight'] < 300:
            H.add_edge(u,v)

    # draw with matplotlib/pylab

    try:
        import matplotlib.pyplot as plt
        plt.figure(figsize=(8,8))
        # with nodes colored by degree sized by population
        node_color=[float(H.degree(v)) for v in H]
        nx.draw(H,G.position,
             node_size=[G.population[v] for v in H],
             node_color=node_color,
             with_labels=False)

        # scale the axes equally
        plt.xlim(-5000,500)
        plt.ylim(-2000,3500)

        plt.savefig("knuth_miles.png")
    except:
        print ('something went wrong')
```

## Dags (workflow charts, etcetera)

```python
Directed acyclic graphs (or DAGS) can be used to represent a myriad of complex relations. 
One of the most important is just to represent dependencies
```

```python
G = nx.DiGraph()

# add 5 nodes, labeled 0-4:
map(G.add_node, range(5))
# 1,2 depend on 0:
G.add_edge(0,1)
G.add_edge(0,2)
# 3 depends on 1,2
G.add_edge(1,3)
G.add_edge(2,3)
# 4 depends on 1
G.add_edge(1,4)

# now draw the graph:
pos = { 0 : (0,0), 1 : (1,1), 2 : (-1,1),
        3 : (0,2), 4 : (2,2)}
lbls = dict((n,n) for n in G.nodes())
```

```python
nx.draw(G, pos, edge_color='orange',labels=lbls,node_size=400)
```

A possible sequence of events for this workflow:

    Task 0 can run right away
    0 finishes, so 1,2 can start
    1 finishes, 3 is still waiting on 2, but 4 can start right away
    2 finishes, and 3 can finally start


```python
from random import randint
def random_dag(nodes, edges):
    """Generate a random Directed Acyclic Graph (DAG) with a given number of nodes and edges."""
    G = nx.DiGraph()
    for i in range(nodes):
        G.add_node(i)
    while edges > 0:
        a = randint(0,nodes-1)
        b=a
        while b==a:
            b = randint(0,nodes-1)
        G.add_edge(a,b)
        if nx.is_directed_acyclic_graph(G):
            edges -= 1
        else:
            # we closed a loop!
            G.remove_edge(a,b)
    return G
```

```python
rg = random_dag(32,128)
nx.draw_spring(rg,node_size=150)
```

## Drawing a network with custom labels

Say you dont like the way networkx is drawing your nodes. When all else fails, 
take charge of labeling your nodes yourself, using matplotlib like the annotate facility just illustrated.
We illustrate using the Les Mis graph used above.  THe time we try for a display in which all the
characters are shown. First draw **only** the edges.

```python
import networkx as nx
import math

lm = nx.read_gml('lesmiserables.gml')
#pos = nx.spring_layout(lm,scale=1.0,iterations=200)
# This is a reasonable default. Assume a square figure. Assume random placement of nodes.  The way to get
# a get a grid with n cells is to have a $\sqrt{n}$ x $\sqrt{n}$ square.
default_distance_between_nodes = 1.0/(math.sqrt(len(lm.nodes())))
#pos = nx.spring_layout(lm,scale=1.0,k=1.2*default_distance_between_nodes) # Try adjusting 
pos = nx.spring_layout(lm)#default distance between nodes
nx.draw_networkx_edges(lm,pos)
```

Then basically draw the nodes yourself, using the *text* function and the node coordinates given
in what is returned by the layout function. But what exactly **is returned** by the layout
function?  A dictionary of coordinates for each node.

```python
pos
```

This is just what we need to make use of the *text* function.  The *text* function places text at some intended location, possibly
with a bounding box.

```python
for label, (x,y) in pos.items():
    plt.text(x, y, label,
             bbox = dict(boxstyle = 'round,pad=0.15', fc = 'salmon', alpha = .5))
```

We iterate through all the nodes and nodes positions in the layout
dictionary. Line 2 opens the call to annotate for a given node, and the next lines specify
the label (line 3), the x,y position of the node, the horizontal and vertical alignment (the last 3 are all on line 5),
and finally what kind of framing box will be placed around the node.  The *alpha* value specifies
how transparent the box will be.  We've picked a value of .5 to make the boxes semi transparent
to help out with those cases where two boxes overlap.
