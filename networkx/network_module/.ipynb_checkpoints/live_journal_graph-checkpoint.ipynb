````python
ucsd_g = res['ucsd']
ucsd_degree_dict = nx.degree(ucsd_g)
````

````python

def sort_score_dict (dd):
    return sorted(dd.items(),key=lambda x:x[1], reverse=True)

def print_score_il (il,score_type=int):
    print 'Max degree: {:d}'.format(len(il))
    print
    for (nd,score) in il:
        if score_type == int:
            print '{0:<20} {1:d}'.format(nd,score)
        else:
            print '{0:<20} {1:.3f}'.format(nd,score)

ucsd_degree_il = sort_score_dict(ucsd_degree_dict)
print_score_il(ucsd_degree_il)
````

````python

sfwa_degree_dict = nx.degree(res['sfwa'])
sfwa_degree_il = sort_score_dict(sfwa_degree_dict)
print_score_il(sfwa_degree_il)
````

````python
sfwa_clustering_dict = nx.clustering(res['sfwa'])
ucsd_clustering_dict = nx.clustering(res['ucsd'])
print 'SFWA'
sfwa_clustering_il = sort_score_dict(sfwa_clustering_dict)
print_score_il(sfwa_clustering_il,score_type=float)
````

This is a HIGHLY interconnected group, as we saw above.  Here it appears we have a strong core of max modularity, but appearances are deceptive.  The highest modularity nodes actually have low degree. This make sense; the higher degree of node `n`, the harder it gets to have a modularity of 1.0.  We show this, and show that even though the highest degree nodes in the `sfwa` graph have modularities below .2, there are some high-degree nodes with modularity above .5.

````python
def print_mod_degree (G):
    clustering_dict = nx.clustering(G)
    degree_dict = nx.degree(G)
    clustering_il = sort_score_dict(clustering_dict)
    degree_il = sort_score_dict(degree_dict)
    print 'Ordered by cluster coeff. Max degree: {:d}'.format(len(degree_dict)-1)
    print
    for (nd,score) in clustering_il:
        print '{0:<20} {1:.3f} {2:d}'.format(nd,score,degree_dict[nd])
    print
    print 'Ordered by degree Max clustering coeff: {0:.3f}'.format(max(clustering_dict.values()))
    print
    for (nd,score) in degree_il:
        print '{0:<20} {1:d} {2:.3f}'.format(nd,score,clustering_dict[nd])

sfwa_g = res['sfwa']
print_mod_degree(sfwa_g)
          
````

## Find communities and color nodes by community

First import the functions defined in the `live_journal_tools` module.

Next choose which graph to find communities for and use the Louvain algorithm to find communities,
returned in the form of a dictionary that assigns a community to each node:
   ```
   >>> partition[n]
   c
   ```
We need a name for each community, but since we know nothing about the constituent nodes besides their position in the graph, we use integers to name the communities found by the algorithm.  If the algorithm finds `k` communities, they will be named 0 through `k-1`.

````python
ucsd_g = res['ucsd']
sfwa_g = res['sfwa']
# UCSD community
#this_g = ucsd_g
# SFWA Community
#this_g = sfwa_g
# Use the whole Live Journal graph !
this_g = lj_graph
#use_louvain = True
use_louvain = False
if use_louvain:
    # compute the best partition via Louvain
    partition = louvain.best_partition(this_g)
else:
    # only use the orgiunal LJ communities
    partition = make_community_partition (this_g,desc_dict)
# How many communities were discovered, altogether?
size = len(set(partition.values()))
print 'We discovered {0:d} partitions linking up {1:d} people!'.format(size,len(lj_graph.nodes()))
````

Now we draw the graph, assigning an arbitrary color to each community.  The following code is written so as to not depend on how many communities were found.  It is a property of `Louvain` that it is hard to predict what this number will be.  The code allows us to choose a subset of the communities, or a subset of the graph, to draw.  We can designate a subgraph either by directly choosing a set of partitions (the strategy used by `find_large_partitions`) or by choosing a set of nodes and then finding the set of partitions it defines (the strategy taken by `find_largest_connected_component_partitions`).

````python
##################################################################
#   Drawing the graph USING the communities discovered by louvain
##################################################################
## Parameters
graph_specific_partitions = True
#graph_specific_partitions = False
figure_width,figure_height = 20,10
node_size = 100
find_large_ps,find_large_component = False, True
#cmap = 'rainbow'
#cmap = 'Spectral'
cmap = 'jet'
#cmap = 'nipy_spectral'
#cmap = 'hsv'


if graph_specific_partitions:
    # Set this by hand
    # For example,The set of ps from the ucsd graph with 9 or more members
    if find_large_ps and not find_large_component:
        chosen_ps = find_large_partitions(partition, threshhold=9)
    elif find_large_component and not find_large_ps:
         # take the largest connected comnponent.
         chosen_ps = find_largest_connected_component_partitions (this_g, partition)
    elif find_large_component and find_large_ps:
        # Or do both
        chosen_ps = find_large_partitions_in_large_component (this_g,partition)
else:
   # If you dont want to pay attention to this
   chosen_ps = set(partition.values())
    #chosen_ps = set([1, 2, 3, 6, 9, 10, 11, 12])


# Graph a fairly random subset of communities.  For example to graph 1 community
# Let mod = size and K = THE COMMUNITY NUMBER
#   k =3 (in the ucsd graph): draws the community containing elefuntboy, the node of highest degree.
#   k = 4 or 20 or 30 or 40: draws a silly community with 2 members
#   k = 50: draws an even sillier "community" with 1 member
# To graph all comunities, use mod=1, k=0
# To graph about 1/3 of the commuities, use mod =3, k=0,1,2
#mod,k = size,6
#mod,k = 3,0
#mod,k = 1,0
#chosen_ps = [n for n in this_g.nodes() if partition[n]%mod==k]

this_g_small, small_partition = filter_graph_and_partition (this_g, partition, chosen_ps)
print 'We draw only the nodes belonging to the largest partitions in the largest component.'
print '{0:d} partitions after filtering!'.format(len(set(small_partition.values())))
# l_color_map {partition=>partition_color}  louvain_colors {node=>node_color}
l_color_map,louvain_colors = get_partition_coloring (this_g_small, small_partition,cmap=cmap)
pos = nx.spring_layout(this_g_small)

#plt.figure(211,figsize=(figure_width,figure_height))
fig,axes = plt.subplots(nrows = 1, ncols=2,figsize=(figure_width,figure_height))
ax0,ax1 = axes
nx.draw_networkx_nodes(this_g_small, pos, this_g_small.nodes(), node_size = node_size,
                                node_color = louvain_colors,ax=ax0)
nx.draw_networkx_edges(this_g_small, pos, alpha=0.5,ax=ax0)
make_partition_color_bar_plot (l_color_map,small_partition,desc_dict,ax=ax1)
#plt.show()
````

## Find cross community links

When we look, we find quite a lot of links across the communities in graphs.  We rank the community pairing by the number of cross community links they have, wand normalize that number by the community sizes.

````python
from collections import Counter,defaultdict
link_ctr,cross_link_ctr,comm_pairs,nds = (0,0,Counter(),Counter())
for (s,e) in lj_graph.edges():
    link_ctr += 1
    s_comm, e_comm = desc_dict[s]['community'], desc_dict[e]['community']
    if s_comm <> e_comm:
        cross_link_ctr += 1
        nds[s] += 1
        nds[e] += 1
        comm_pairs[tuple(sorted((s_comm,e_comm)))] += 1
ind_comm = defaultdict(list)
for (k,g) in res.iteritems():
    for nd in g.nodes():
        ind_comm[nd].append(k)
# Print out an ordered list of who has cross community links.  Show their home community/
for (nd,ct) in sorted(nds.iteritems(),reverse=True,key = lambda x:x[1]):
    print '{0:<20} {1:<3d} {2}'.format(nd,ct,ind_comm[nd])
````

````python
# Assumes  `res` as defined in the first cell in this NB is available.

def print_graph_sizes (res):
    """
    Prints and returns a dictionary that gives the size of the subgraph belonging to each of the
    communities.
    """
    size_dict = dict()
    for k in sorted(res.keys(),reverse=True,key=lambda x: len(res[x].nodes())):
        size = len(res[k].nodes())
        print '{0:25} {1:d}'.format(k,size)
        size_dict[k] = size
    return size_dict
    
def print_cross_comm_stats (comm_pairs,size_dict):
    """
    Prints two statistics about cross community connections:  (a) The count
    of teh number of connections between community pairs and
    (b) that count normalized by the number of possible connections
    given the number of members in the two communities.
    
    According to this normalized number the most interesting cross
    community connection in the LJ data is between `compling`
    and `multiple`.
    """
    il = sorted(comm_pairs.iteritems(), reverse=True, key = lambda x:x[1])
    for (i,((st,en), ct)) in enumerate(il):
        # Normalize the connection counts by the group sizes
        norm = (size_dict[st] * size_dict[en])/2.0
        print '{0:>2d}. {1:<15}  {2:<15} {3: 3d} {4:.3f}'.format(i+1,st,en,ct,float(ct)/norm)

print 
size_dict = print_graph_sizes(res)
print
print_cross_comm_stats (comm_pairs,size_dict)
````
